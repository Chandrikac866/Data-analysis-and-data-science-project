```python

gen=sns.countplot(x="gender",data=data,color="yellow")
gen.bar_label(gen.containers[0])
plt.title("GENDER BASED ON DATA")
plt.show
```

```python
month=sns.histplot(x="monthly_fee",data=data)
month.bar_label(month.containers[0])
plt.title(" totals Monthly_fees present in data ")
plt.show()
```

```python
pm=sns.distplot(data["churned"])
pm.bar_label(pm.containers[0])

plt.title("Total number of churned data ")
plt.show()
```

```python
sns.boxplot(x="last_login_days",data=data,gap=0)
plt.title("last_login_days box plot")
plt.show()
```

```python
sea=sns.scatterplot(x="watch_hours" ,y="device",data=data,legend="auto",color="purple",)
plt.title("Scatter plot using watch_hours and device")
plt.show()
```

```python
sns.lineplot(y="payment_method",x="region",data=data,)
plt.title("Payment method vs region")
plt.show()
```
